```python
import numpy as np 
from keras.preprocessing.image import image
from keras.preprocessing.image import img_to_array
from keras.applications.resnet50 import preprocess_input
from keras.applications.resnet50 import ResNet50,decode_predictions
import matplotlib.pyplot as plt

```

```python
# preprocessing images for ResNet50
# When loading the picture remenber to set the right target size which for ResNet is 224*224.
img = image.load_img('dog.jpg',target_size = (224,224))
plt.imshow(img)
```

```python
'''Keras is expecting a list of images which is why we need to turn it into an array and then add another dimension to this. Following this we need to normalize the image using preprocess input method.'''
```

```python
img = image.img_to_array(img)
img.shape
```

```python
img = np.expand_dims(img,axis=0)
img.shape
```

```python
# preprocessing the numpy array encoding a batch of images
img = preprocess_input(img)
```

```python
#make a prediction
model = ResNet50(weights='imagenet')
```

```python
preds = model.predict(img)
preds.shape
```

```python
print(decode_predictions(preds,top=1)[0])
```
